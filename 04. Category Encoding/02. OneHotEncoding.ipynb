#### 1. 데이터 호출

```python
import pandas as pd

train = pd.read_csv("data/income_train.csv")
test = pd.read_csv('data/income_test.csv')
```

#### 2. train 데이터에 OneHotEncoding 적용

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# train 복제본 생성
train_ex1 = train.copy()

# 범주형 변수 설정
categorical_var = ['workclass','education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']

# OneHotEncoding 적용
encoder = OneHotEncoder(sparse_output=False)
train_encoded = encoder.fit_transform(train_ex1[categorical_var])

# OneHotEncoding 적용한 DataFrame 만들기
encoded_columns = encoder.get_feature_names_out(categorical_var)
train_encoded_df = pd.DataFrame(train_encoded, columns=encoded_columns)

# train과 train_encoded_df 병합
train_ex1 = pd.concat([train_ex1, train_encoded_df], axis=1).drop(categorical_var, axis=1)

display(encoded_columns)
display(train)
```

#### 3. test 데이터에 적용하기

```python
test_ex1 = test.copy()

# test 데이터에 OneHotEncoding 적용
test_encoded = encoder.transform(test_ex1[categorical_var])

# 추출한 Feature를 해당 Column에 적용
test_encoded_df = pd.DataFrame(test_encoded, columns = encoded_columns)

# OneHotEncoding된 데이터와 test 데이터 병합 후 기존 컬럼 제거
test_ex1 = pd.concat([test_ex1, test_encoded_df], axis=1).drop(categorical_var, axis=1)

display(test_ex1)
```

#### 4. test 데이터에 없는 값 처리하기

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# train, test 데이터 복제
train_ex2 = train.copy()
test_ex2 = test.copy()

# 범주형 변수 설정
categorical_var = ['workclass','education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']

# OneHotEncoding에서 train 데이터에 없는 값 무시
# 이 조건을 설정할 경우 해당 값은 모두 0이 됨
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# train 데이터에 대해 OneHotEncoding 적용
train_encoded = encoder.fit_transform(train_ex2[categorical_var])

# OneHotEncoding 된 Column 추출
encoded_columns = encoder.get_feature_names_out(categorical_var)

# array를 DataFrame으로 변경 후 컬럼 설정
train_encoded_df = pd.DataFrame(train_encoded, columns=encoded_columns)

# train 데이터와 병합
train_ex2 = pd.concat([train_ex2, train_encoded_df], axis=1).drop(categorical_var, axis=1)

# test 데이터에 OneHotEncoding 적용
test_encoded = encoder.transform(test_ex2[categorical_var])

# test 데이터 DataFrame으로 변환
test_encoded_df = pd.DataFrame(test_encoded, columns = encoded_columns)

# test 데이터와 병합
test_ex2 = pd.concat([test_ex2, test_encoded_df], axis=1).drop(categorical_var, axis=1)

display(train_ex2.shape)
display(test_ex2.shape)
```

#### 5. Dummy 변수의 함정과 OneHotEncoding의 Drop 옵션

###### [1] dummy 변수로 하나의 컬럼을 여러개로 나눈다면, 해당 컬럼들 사이에는 연관성이 존재한다. 

###### [2] 예를들어 'Brand_Ford'가 0이라면, 이는 자동적으로 'Brand_Honda' 또는 'Brand_Toyota' 중 하나가 1이 될 가능성이 높다는 것을 의미한다.

###### [3] 즉 해당 컬럼들 사이에는 다중공선성이 형성된다는 것이다.

###### [4] 이러한 문제는 'Dummy 변수의 함정'이라고 불린다.

###### [5] 이를 해결하기 위해 sklearn에서는 OneHotEncoding의 drop 매개변수를 사용한다.

###### [6] drop = 'first'로 설정하면, 첫번째 컬럼을 자동으로 삭제하여 다중공선성을 줄이면서 필요한 정보를 포함시킬 수 있다.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# 데이터 복제
train_ex3 = train.copy()
test_ex3 = test.copy()

# OneHotEncoder 불러오기
encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')

# 범주형 변수 설정
categorical_var = ['workclass','education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']

# train 데이터에 OneHotEncoding 적용
train_encoded = encoder.fit_transform(train_ex3[categorical_var])

# OneHotEncoding된 Columns 추출
encoded_columns = encoder.get_feature_names_out(categorical_var)

# 해당 변수를 DataFrame으로 만들기
train_encoded_df = pd.DataFrame(train_encoded, columns=encoded_columns)

# train 데이터와 병합
train_ex3 = pd.concat([train_ex3, train_encoded_df], axis=1).drop(categorical_var, axis=1)

# test 데이터 OneHotEncoding 적용
test_encoded = encoder.transform(test_ex3[categorical_var])

# OneHotEncoding된 값을 DataFrame으로 변환
test_encoded_df = pd.DataFrame(test_encoded, columns = encoded_columns)

# test 데이터와 encoding된 DataFrame 결합
test_ex3  = pd.concat([test_ex3, test_encoded_df], axis=1).drop(categorical_var, axis=1)

display(test_ex3)
```
